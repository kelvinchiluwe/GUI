```python
df = pd.read_csv('company_data.csv')
df['Order Date'] = pd.to_datetime(df['Order Date'], dayfirst=False, errors='coerce')
df['Month'] = df['Order Date'].dt.strftime('%B')
kelvin=df.head(30)

order = kelvin['Month']
sales = kelvin['Sales']
profit = kelvin['Profit']
Qty = kelvin['Order Qty']
plt.subplots_adjust(hspace=0.9)
fig, axs = plt.subplots(3, 1, figsize=(8, 10), constrained_layout=True) 

axs[0].plot(order, sales, color='blue', markerfacecolor='g', linestyle='--')
axs[0].set(title='Order Date vs Sales', xlabel='Order Date', ylabel='Sales (mkw)')
plt.xticks(rotation=45)
axs[0].grid(True)

axs[1].bar(order, profit)
axs[1].set(title='Order Date vs Profit', xlabel='Order Date', ylabel='Profit (mkw)')
axs[1].grid(True)

axs[2].hist(order, bins=10, color='red', alpha=0.7)
axs[2].set(title='Order Date vs Order Qty', xlabel='Order Date', ylabel='Order Qty')
axs[2].grid(True)

```
